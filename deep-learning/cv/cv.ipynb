```python
# !pip install opencv-python

import cv2

image = cv2.imread("./monkey-d-luffy-3840x2160-16712.jpg")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')

# plt.show()

    
```

```python
# crop image
cropped_image = image[0:1550,1000:1500]
plt.imshow(cropped_image)
# cv2.imwrite("new.jpg", cropped_image)
```

```python
mirrored_image = cv2.flip(image, 0)  
upwards_image = cv2.flip(mirrored_image, 0)

plt.imshow(upwards_image)
```

```python
# crop image
transpose = cv2.transpose(upwards_image)
plt.imshow(transpose)
```

```python
flipped_h = cv2.flip(transpose, -1) 
plt.imshow(flipped_h)
```
